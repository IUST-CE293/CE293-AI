و نعیین مقدارش خیلی مهمه چون اگه خیلی بزرگ انتخاب بشه، مدل دچار اورفیت میشه و اگه خیلی کم انتخاب بشه مدل دچار آندرفیت میشه(راجب این تحقیق کنید که چرا اینطوریه به نظرتون؟؟)

هایپرپارامتر مثل تنظیمات یه دستگاه هستن که باید قبل از کار با اون دستگاه تنظیم بشن، مثلا یک تردمیل رو در نظر بگیری، سرعت حرکت اون تردمیل

میشه یک هایپرپارامتر. حالا علمی تر بخوایم بگیم، هایپرپارامتر تنظیمات مدل هست که قبل از train مدل باید اعمال بشه.

یا مثلا باز توی درخت تصمیم، min_samples_split  یک هایپرپارامتره.

حالا ما یک فرآیندی داریم به اسم hyperparameter tuning که با دیتاست validation قراره انجام بشه و به صورت کلی به این شکله که

ما قراره یک بهینه سازی یا optimization انجام بدیم تا بهینه ترین هایپرپارامتر ها برای مدل انتخاب بشن.

یکی از روش هایی تقسیم دیتا برای hyperparameter tuning روش held out هست که در این روش شما اول میای مثلا 20 درصد دیتا رو از اول میذاری کنار برای تست؛

سپس از 80 درصد باقی مونده، دوباره 80درصدشو(64درصد دیتای اولیه) میذاری برای train و مابقی رو برای validation قرار میدهی

حالا با این دیتاهای باقی مونده ی validation , hyperparameter tuning انجام میدیم تا بهینه ترین هایپرپارامتر ها رو انتخاب کنیم،

بعد که کارمون تموم شد، دیتاست train , validation رو میریزیم روی هم و اینا رو در مجموع میدیم به مدل برای train.
